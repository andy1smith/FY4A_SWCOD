```python
def relative_zenith_azimuth_angle1(xyzr_cld, xy_obs):
    x_cld, y_cld, ztop_cld, zbotm_cld, r_cld = xyzr_cld
    x_obs, y_obs, z_obs = xy_obs

    if (x_cld-r_cld)*(x_cld+r_cld)<=0 and (y_cld-r_cld)*(y_cld+r_cld)<=0:  # right above the head
        phi_range = np.array([0, 2 * math.pi])

        Ed = math.sqrt(((x_cld - r_cld) - x_obs) ** 2 + ((y_cld - r_cld) - y_obs) ** 2 + (ztop_cld - z_obs) ** 2)
        D = math.sqrt(((x_cld - r_cld) - x_obs) ** 2 + ((y_cld - r_cld) - y_obs) ** 2) + r_cld

    else:  # not right above the head
        # Calculate the realtive zenith angle
        Ed = math.sqrt(((x_cld - r_cld) - x_obs) ** 2 + ((y_cld - r_cld) - y_obs) ** 2 + (ztop_cld - z_obs) ** 2)
        theta1 = math.acos((ztop_cld - z_obs) / Ed)  # rad
        Ed = math.sqrt(((x_cld + r_cld) - x_obs) ** 2 + ((y_cld + r_cld) - y_obs) ** 2 + (zbotm_cld - z_obs) ** 2)
        theta2 = math.acos((zbotm_cld - z_obs) / Ed)  # rad
        theta_range = np.array([theta1, theta2])
        # Calculate the azimuth angle to the center of the cloud
        phi_center = math.atan2(y_cld - y_obs, x_cld - x_obs)  # (-pi, pi)
        Ed = math.sqrt((x_cld - x_obs) ** 2 + (y_cld - y_obs) ** 2)
        phi_r = math.asin(r_cld / Ed)
        phi_range = np.array([phi_center - phi_r, phi_center + phi_r])

    return theta_range, phi_range
```

```python
def create_gif(image_list, gif_name, duration=5):
    frames = []
    for image_name in image_list:
        frames.append(imageio.imread(image_name))
    imageio.mimsave(gif_name, frames, 'GIF', duration=duration)
    return
```

```python
from PIL import Image
import imageio

def create_gif(image_list, gif_name, duration):
    frames = []
    # Load the first image to get size
    base_image = Image.open(image_list[0])
    base_size = base_image.size  # (width, height)

    for image_name in image_list:
        img = Image.open(image_name).convert('RGB')
        img_resized = img.resize(base_size)
        frames.append(img_resized)

    frames[0].save(
        gif_name,
        format='GIF',
        append_images=frames[1:],
        save_all=True,
        duration=duration,
        loop=0
    )
```
