# Download HITRAN data

```python
# only need to run once, save data into local directory data/hitran_LW or data/hitran_SW
from LBL_funcs_getHitran import *
molecules=['CO']#['H2O','CO2','O3','N2O','CH4','O2','N2'] # seven main participating gases in the atmosphere
```

```python
# download longwave HITRAN data
nu_min=0
nu_max=2500
spectral='LW'
for i in range(0,len(molecules)):
    getDataFromHitran(molecules[i], nu_min, nu_max, spectral)  
```

```python
# download shortwave HITRAN data
nu_min=2500
nu_max=35000 # > 10^4/0.29 um
spectral='SW'
for i in range(0,len(molecules)):
    getDataFromHitran(molecules[i], nu_min, nu_max, spectral)  
```
