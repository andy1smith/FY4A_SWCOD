```python
import pandas as pd
import os

dirpath = './'

channels = ['C01', 'C02', 'C03', 'C04', 'C05', 'C06'] 

for channel in channels:
    channel_number = int(channel[-2:])
    input_file = os.path.join(
        dirpath,
        'FY4A_AGRI_SRF_ch{:d}_20180306.txt'.format(channel_number)
    )

    output_file = os.path.join(
        dirpath,
        'FY4A_AGRI_SRF_ch{:d}.txt'.format(channel_number)
    )
    try:
        df = pd.read_csv(input_file, sep='\s+')
        df = df.rename(columns={
                            'nm': 'wavelength/nm',
                            '%': 'RSRF/%'
                            })
        df['wavenumber/cm-1'] = 10**7 / df['wavelength/nm']
        df['RSRF/1'] = df['RSRF/%']/100

        new_column_order = ['wavelength/nm', 'wavenumber/cm-1', 'RSRF/1']
        df = df[new_column_order]

        df.to_csv(output_file, index=False)

        print(f"Success! File processed and saved to '{output_file}'")

    except FileNotFoundError:
        print(f"Error: Input file '{input_file}' not found.")
    except KeyError as e:
        print(f"Error: A column header {e} was not found in your file.")
        print("Please check that your .txt file headers match the code.")
    except Exception as e:
        print(f"An error occurred: {e}")
```
